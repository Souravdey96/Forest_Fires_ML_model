```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
```

```python
df.head()
```

```python
df.info()
```

# **Data Cleaning**

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122, "Region"]=0
df.loc[122:, "Region"]=1
```

```python
df.info()
```

```python
df[['Region']] = df[['Region']].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
## Removing the Null Values
df = df.dropna().reset_index(drop=True)
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
# Removed the 122nd Row
df = df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
# Fix spaces in column names
df.columns = df.columns.str.strip()
df.columns
```

```python
df.info()
```

## **Change the required columns as integer datatype**

```python
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']] = df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

#### Changing the other columns to float datatype

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
  if i!='Classes':
    df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
## let save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```

```python
#plotting pie chart
percentage = df_copy['Classes'].value_counts(normalize=True)*100
classLabels = ["Fire", "Not Fire"]
plt.figure(figsize=(12, 7))
plt.pie(percentage, labels=classLabels, autopct='%1.1f%%')
plt.title("Pie chart of Classes")
plt.show()
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'), 'not fire', 'fire')
```

```python
dftemp=df.loc[df['Region']==1]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')

# Define a custom palette for 'Classes'
custom_palette = {'fire': 'black', 'not fire': 'green'}

sns.countplot(x='month', hue='Classes', data=df, palette=custom_palette)
plt.title('Fire Analysis of Sidibel Regions', weight='bold')
plt.xlabel('Months', weight='bold')
plt.ylabel('Number of Fires', weight='bold')
plt.legend(title='Classes')
plt.show()
```

```python
dftemp=df.loc[df['Region']==0]
plt.subplots(figsize=(13,6))
sns.set_style('whitegrid')

# Define a custom palette for 'Classes'
custom_palette = {'fire': 'black', 'not fire': 'green'}

sns.countplot(x='month', hue='Classes', data=df, palette=custom_palette)
plt.title('Fire Analysis of Bejaia Regions', weight='bold')
plt.xlabel('Months', weight='bold')
plt.ylabel('Number of Fires', weight='bold')
plt.legend(title='Classes')
plt.show()
```
